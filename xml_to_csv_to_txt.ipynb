```python
from xml.etree import ElementTree
import os
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount("/content/drive")

# 請輸入到frcnn資料夾的路徑

path = "/content/drive/MyDrive/Colab Notebooks/frcnn"
os.chdir(path)

# Define columns and rows
cols = ["filename", "type", "xmin", "xmax", "ymin", "ymax"]
rows = []

# Define the path to the XML files
xml_folder_path = os.path.join(path, 'xml')

# Check if the XML folder path exists
if not os.path.exists(xml_folder_path):
    raise FileNotFoundError(f"The folder {xml_folder_path} does not exist.")

# Iterate over XML files in the folder
for filename in os.listdir(xml_folder_path):
    file_path = os.path.join(xml_folder_path, filename)
    if os.path.isfile(file_path) and filename.endswith(".xml"):
        print(f"Processing file: {file_path}")
        try:
            with open(file_path, encoding='utf-8') as file:
                tree = ElementTree.parse(file)
                root = tree.getroot()

                # Extract filename
                img_filename = root.find('filename').text

                # Extract object details
                for obj in root.findall("object"):
                    name = obj.find("name").text
                    bndbox = obj.find("bndbox")
                    xmin = bndbox.find("xmin").text
                    xmax = bndbox.find("xmax").text
                    ymin = bndbox.find("ymin").text
                    ymax = bndbox.find("ymax").text

                    # Append row to the list
                    rows.append({
                        "filename": img_filename,
                        "type": name,
                        "xmin": xmin,
                        "xmax": xmax,
                        "ymin": ymin,
                        "ymax": ymax
                    })
        except ElementTree.ParseError as e:
            print(f"Error parsing {file_path}: {e}")
        except Exception as e:
            print(f"An error occurred with {file_path}: {e}")

# Create DataFrame
df = pd.DataFrame(rows, columns=cols)

# Save DataFrame to CSV
output_csv_path = os.path.join(path, 'annotations.csv')
df.to_csv(output_csv_path, index=False)

print(f"CSV file saved to: {output_csv_path}")
```

```python
import pandas as pd

# # 請輸入到annotations的路徑
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/frcnn/annotations.csv')

# Filter the DataFrame to only include filenames from 0.jpg to 79.jpg
filtered_train = train[train['filename'].isin([f'{i}.jpg' for i in range(80)])]

# Create a new DataFrame for formatted data
data = pd.DataFrame()
data['format'] = filtered_train['filename']

# Add 'test_images/' before each filename
data['format'] = 'test_images/' + data['format']

# Add xmin, ymin, xmax, ymax, and class as per the format required
for i in range(data.shape[0]):
    row = filtered_train.iloc[i]
    data['format'].iloc[i] = f"{data['format'].iloc[i]},{row['xmin']},{row['ymin']},{row['xmax']},{row['ymax']},{row['type']}"

# Save the data to 'annotate.txt'
data.to_csv('annotate.txt', header=None, index=None, sep=' ')

print("Annotations saved to 'annotate.txt'")
```
